# Matplotlib: A Fundamental Data Visualization Library
Matplotlib is the foundational plotting library for Python, offering a comprehensive set of tools for creating static, animated, and interactive visualizations.

# Key Features
* Publication-quality figures in various formats (PNG, PDF, SVG, etc.)
* Fine-grained control over plot elements
* Works seamlessly with NumPy and Pandas
* Supports diverse plot types: line plots, scatter plots, histograms, bar charts, etc.

# Common Use Cases in Data Science
* Exploring data distributions and patterns
* Visualizing trends and relationships
* Creating statistical plots
* Building complex data dashboards
* Time series analysis visualization
* Model performance evaluation plots

# Example
The code in this notebook demonstrates a basic sine wave plot using Matplotlib, showcasing:

* Figure size customization
* Line plotting with color specification
* Title and axis labels
* Grid and legend implementation

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='sin(x)')
plt.title('Basic Matplotlib Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [3, 2, 6, 8, 10]

# Set figure size and resolution
plt.figure(figsize=(6, 4), dpi=80)

# Plot first line with points and solid line
plt.plot(x, y, 'r.-', label="2x")

# Create more precise x values using numpy
x2 = np.arange(0, 4.5, 0.5)

# Plot x^3 function in two parts with different styles
plt.plot(x2[4:], x2[4:]**3, 'b*--', label='x^3')  # Dashed line with stars
plt.plot(x2[0:5], x2[0:5]**3, 'g.-', label='x^3')  # Solid line with points

# Set title with a standard font
plt.title("My first graph!", fontsize=20)

# Set axis labels
plt.xlabel("x axis")
plt.ylabel("y axis")

# Set x-axis ticks
plt.xticks([0, 1, 2, 3, 4])

# Add legend
plt.legend()

# Save figure with high resolution
plt.savefig('mygraph.png', dpi=200)

# Display the plot
plt.show()
```

```python
labels =['A','B','C']
values =[1,4,3]
plt.title("My first Bar Charts!",fontdict={"fontsize":20})

bars=plt.bar(labels,values,)
plt.figure(figsize=(6,4))
# bars[0].set_hatch('+')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

# using loop instead of above code
patterns= ['+','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('mycharts.png',dpi=300)

    
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('people_data.csv')
```

```python
# Plot: Age distribution histogram
plt.figure(figsize=(15, 10))

plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

plt.show()
```

```python
# Convert salary strings to numeric values
salary_numeric = df['salary'].str.replace('k', '').astype(float)

plt.figure(figsize=(10, 6))
plt.scatter(df['age'], salary_numeric, color='blue', alpha=0.6)
plt.plot(df['age'], salary_numeric, linestyle='-', color='red', alpha=0.3)
plt.title('Age vs Salary Trend')
plt.xlabel('Age')
plt.ylabel('Salary (k)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
city_count = df['city'].value_counts()
plt.pie(city_count.values, labels=city_count.index, autopct='%1.1f%%')
plt.title('Employee Distribution by City')

plt.show()
```
